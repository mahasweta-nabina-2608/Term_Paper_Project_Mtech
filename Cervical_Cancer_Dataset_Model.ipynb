```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'kag_risk_factors_cervical_cancer.csv'
df = pd.read_csv(C:\Users\KIIT\Desktop\Genetic Diseases Dataset)

# Data Cleaning: Replace '?' with NaN and convert columns to numeric where possible
df.replace('?', np.nan, inplace=True)
df = df.apply(pd.to_numeric, errors='ignore')

# Bar plot: Distribution of cancer diagnosis (Dx:Cancer) in the dataset
plt.figure(figsize=(8, 6))
sns.countplot(x='Dx:Cancer', data=df)
plt.title('Distribution of Cancer Diagnosis')
plt.xlabel('Cancer Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Scatter plot: Relationship between Age and Number of Pregnancies
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Num of pregnancies', data=df)
plt.title('Age vs. Number of Pregnancies')
plt.xlabel('Age')
plt.ylabel('Number of Pregnancies')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\KIIT\Desktop\Genetic Diseases Dataset\kag_risk_factors_cervical_cancer.csv'
df = pd.read_csv(file_path)

# Data Cleaning: Replace '?' with NaN and convert columns to numeric where possible
df.replace('?', np.nan, inplace=True)
df = df.apply(pd.to_numeric, errors='ignore')

# Bar plot: Distribution of cancer diagnosis (Dx:Cancer) in the dataset
plt.figure(figsize=(8, 6))
sns.countplot(x='Dx:Cancer', data=df)
plt.title('Distribution of Cancer Diagnosis')
plt.xlabel('Cancer Diagnosis (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Scatter plot: Relationship between Age and Number of Pregnancies
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Num of pregnancies', data=df)
plt.title('Age vs. Number of Pregnancies')
plt.xlabel('Age')
plt.ylabel('Number of Pregnancies')
plt.show()
```
