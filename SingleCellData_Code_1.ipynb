```python
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Baron.csv', index_col=0)

# Create an AnnData object (Scanpy's data structure)
adata = sc.AnnData(data.iloc[:, :-2])  # Exclude 'label' and 'sampleID' columns
adata.obs['label'] = data['label']  # Add the labels to the AnnData object

# 1. Quality Control (Remove low-expressed genes or cells)
sc.pp.calculate_qc_metrics(adata, inplace=True)
sc.pl.violin(adata, ['n_genes_by_counts', 'total_counts', 'pct_counts_mt'], jitter=0.4)

# Filter out cells with too few genes or high mitochondrial content
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)

# 2. Normalize the data
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# 3. Identify highly variable genes
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var.highly_variable]

# 4. Perform PCA
sc.tl.pca(adata)
sc.pl.pca_variance_ratio(adata, log=True)

# 5. Compute UMAP for visualization
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['label'])

# 6. Clustering
sc.tl.leiden(adata)
sc.pl.umap(adata, color=['leiden'])

# 7. Identify marker genes
sc.tl.rank_genes_groups(adata, 'leiden', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)

# Save the processed data
adata.write('processed_Baron_data.h5ad')
```
